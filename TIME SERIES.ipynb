# TIME SERIES (Data Science)


Time series is a collection of data which is arranged according to the time.
Because of time series we can

~analysis the all past data .

~collect the present data in precision manner.

~predict the future data output correspondence with the proper  time.


# TYPES OF FACTORS OF A TIME SERIES:

There are some factors which affect the data accordance with the time.
These are called the time series.


•	LONG TERM OR SECULAR OR SIMPLE VARIATION:
Only having a particular trends like increasing of data or decreasing of data
eg. Population (increasing continuously according to the time)


•	SEASONAL VARIATION:
Having a various trends in same year accordance to the month
eg. various seasonal fruits demand


•	CYCLIC VARIATION:
When the data having the phases like:
1.	prosperity
2.	recession
3.	depression
4.	recovery
in cyclic manner

•	RANDOM OR IRREGULAR VARIATION:
Variation taken place due some random problems called as irregular or random variation.
eg. Increment in supplies due to earthquake at the certain places.


# TYPES OF THE MODELS OF THE TIME SERIES ANALYSIS:

•	ADDITIVE MODEL:
           If the various factors are independent with each other than
           Output=Sum of all factors.

•	MULTIPLICATIVE MODEL:
             If the various factors are dependent with each other than
            Output=Multiplication of all factors

# •	MOVING AVERAGES METHODS:
Use MA terms in the model when the model is:

Negatively Autocorrelated at Lag — 1

ACF that drops sharply after a few lags

PACF decreases more gradually


```python
import pandas as pd
from pandas import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
k = pd.read_csv('PRICE_AND_DEMAND_201301_NSW1.csv',parse_dates=['SETTLEMENTDATE'],index_col='SETTLEMENTDATE')
k.head()
```

```python
y=np.array(k['TOTALDEMAND'])
y
```

```python
weights=[]
smas=[]
```

```python
def movingaverage(values,window):
    weights=np.repeat(1.0,window)/window
    smas=np.convolve(values,weights,'valid')
    return smas
```

```python
h=pd.DataFrame(movingaverage(y,3)).set_index(k.index[2:])
k['MA']=h.shift(9)
k.head()
```

```python
plt.figure(figsize=(15,10))
plt.plot(k['TOTALDEMAND'])
plt.plot(k['MA'],color='red')
```

# •	RNN MODEL:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('PRICE_AND_DEMAND_201301_NSW1.csv',parse_dates=['SETTLEMENTDATE'])
df.head()
```

```python
train=df.iloc[:,2:3].values
```

```python
from sklearn.preprocessing import MinMaxScaler  
scaler = MinMaxScaler(feature_range = (0, 1))

train_scaled = scaler.fit_transform(train)
```

```python
features_set = []  
labels = []  
for i in range(48, 1488):  
    features_set.append(train_scaled[i-48:i, 0])
    labels.append(train_scaled[i, 0])
```

```python
features_set, labels = np.array(features_set), np.array(labels)
```

```python
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))
```

```python
from keras.models import Sequential  
from keras.layers import Dense  
from keras.layers import LSTM  
from keras.layers import Dropout
```

```python
model = Sequential()
```

```python
model.add(LSTM(units=50, return_sequences=True, input_shape=(features_set.shape[1], 1)))
```

```python
model.add(Dropout(0.1))
```

```python
model.add(LSTM(units=48, return_sequences=True))  
model.add(Dropout(0.1))

model.add(LSTM(units=48, return_sequences=True))  
model.add(Dropout(0.1))

model.add(LSTM(units=48))  
model.add(Dropout(0.1))
```

```python
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=['accuracy'])
```

```python
model.fit(features_set, labels, epochs = 5, batch_size = 48)
```

```python
predictions = model.predict(features_set)
predictions
```

```python
predictions = scaler.inverse_transform(predictions)
predictions
```

```python
plt.figure(figsize=(10,6))  
plt.plot(train, color='blue', label='Actual TOTALDEMAND')  
plt.plot(predictions , color='red', label='Predicted TOTALDEMAND')  
plt.title('TOTALDEMAND Prediction')  
plt.xlabel('Date')  
plt.ylabel('TOTALDEMAND')  
plt.legend()  
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('PRICE_AND_DEMAND_201301_NSW1.csv',parse_dates=['SETTLEMENTDATE'])
df.head()
```

```python
train=df.iloc[:,2:3].values
```

```python
from sklearn.preprocessing import MinMaxScaler  
scaler = MinMaxScaler(feature_range = (0, 1))

train_scaled = scaler.fit_transform(train)
```

```python
features_set = []  
labels = []  
for i in range(60, 1488):  
    features_set.append(train_scaled[i-60:i, 0])
    labels.append(train_scaled[i, 0])
```

```python
features_set, labels = np.array(features_set), np.array(labels)
```

```python
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1]))
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(activation='relu',units=60,kernel_initializer='uniform',input_dim=60))
classifier.add(Dense(activation='relu',units=60,kernel_initializer='uniform'))
classifier.add(Dense(activation='sigmoid',units=1,kernel_initializer='uniform'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(features_set, labels,batch_size=60,epochs=5) 
```

```python
y_predicted=pd.DataFrame(classifier.predict(features_set))
y_predicted.columns=['y_predicted']
y_predicted.head(10)
```

```python
y_predicted = scaler.inverse_transform(y_predicted)
y_predicted
```

```python
plt.figure(figsize=(10,6))  
plt.plot(train, color='blue', label='Actual TOTALDEMAND')  
plt.plot( y_predicted, color='red', label='Predicted TOTALDEMAND')  
plt.title('TOTALDEMAND Prediction')  
plt.xlabel('Date')  
plt.ylabel('TOTALDEMAND')  
plt.legend()  
plt.show()
```

```python
frame_count = 0
start_time = time.time()

while True:
    

    data = stream.read(CHUNK)  
    
    
    data_int = struct.unpack(str(2 * CHUNK) + 'B', data)

    data_np = np.array(data_int, dtype='b')[::2] + 128
    
    line.set_ydata(data_np)
    

    try:
        fig.canvas.draw()
        fig.canvas.flush_events()
        frame_count += 1
        
    except TclError:
        
        
        frame_rate = frame_count / (time.time() - start_time)
        
        print('stream stopped')
        print('average frame rate = {:.0f} FPS'.format(frame_rate))
        break
        
```

```python
import pyaudio
import os
import struct
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import time
from tkinter import TclError
```

```python
CHUNK = 1024 * 2            
FORMAT = pyaudio.paInt16     
CHANNELS = 1                 
RATE = 44100
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(15, 7))
```

```python
p = pyaudio.PyAudio()
```

```python
stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    frames_per_buffer=CHUNK
)
```

```python
x = np.arange(0, 2 * CHUNK, 2)      
xf = np.linspace(0, RATE, CHUNK) 
```

```python
line, = ax1.plot(x, np.random.rand(CHUNK), '-', lw=2)
line_fft, = ax2.semilogx(xf, np.random.rand(CHUNK), '-', lw=2)
```

```python
ax1.set_title('AUDIO WAVEFORM')
ax1.set_xlabel('samples')
ax1.set_ylabel('volume')
ax1.set_ylim(0, 255)
ax1.set_xlim(0, 2 * CHUNK)
plt.setp(ax1, xticks=[0, CHUNK, 2 * CHUNK], yticks=[0, 128, 255])
```

```python
ax2.set_xlim(20, RATE / 2)

print('stream started')
```

```python
frame_count = 0
start_time = time.time()
```

```python
while True:
    
    
    data = stream.read(CHUNK)  
    
    
    data_int = struct.unpack(str(2 * CHUNK) + 'B', data)
    
    
    data_np = np.array(data_int, dtype='b')[::2] + 128
    
    line.set_ydata(data_np)
    
    
    yf = fft(data_int)
    line_fft.set_ydata(np.abs(yf[0:CHUNK])  / (128 * CHUNK))
    
    
    try:
        fig.canvas.draw()
        fig.canvas.flush_events()
        frame_count += 1
        
    except TclError:
        
    
        frame_rate = frame_count / (time.time() - start_time)
        
        print('stream stopped')
        print('average frame rate = {:.0f} FPS'.format(frame_rate))
        break
```
